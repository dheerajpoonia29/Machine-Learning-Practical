```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("headbrain.csv")
print(data)
```

```python
# shape is attribute of class read_csv
print(data.shape)
```

```python
# shape is method of class read_csv
print(data.head())
```

```python
#IMPLEMENTATION OF LINEAR REGREESION
# x is x-axis value where it is independent variable
#step1: preparation of data set
X = data[["Head Size(cm^3)"]].values
print(X)
```

```python
Y = data[["Brain Weight(grams)"]].values
print(Y)
```

```python
#step2: plot graph b/w y & x
```

```python
#step3: finding simple linear regression (finding mean of x & y)
mean_X = np.mean(X)
print(mean_X)
mean_Y = np.mean(Y)
print(mean_Y)
```

```python
n = len(X)
print("x-axis",n)
n = len(Y)
print("y-axis",n)
```

```python

m = 0
i=0
while(i<n):
    #print(X[i],Y[i])
    m+=(X[i]-mean_X)*(Y[i]-mean_Y)/(X[i]-mean_X)
    i+=1
print(m)
```

```python
#step4: estimating the slope
numerator = 0
denominator = 0
for i in range(n):
    numerator+=(X[i]-mean_X)*(Y[i]-mean_Y)
    denominator+=(X[i]-mean_X)**2
m = numerator/denominator
print(m)

#step5: estimating the intercept
c = (mean_Y - m*mean_X)
print(c)
```

```python
#step6: makind prediction
# for drawing perfect graph
max_X = np.max(X)+100
min_X = np.min(X)-100
print(max_X)
print(min_X)
```

```python
x = np.linspace(min_X, max_X, 1000)
y = m*x + c
# plotting a line (line of regression) or prediction line (x,y)
plt.plot(x,y, color='Green', label='Regression Line')
# plotting data point on line, where X is feature which we show on the line (x,y)
plt.rcParams['figure.figsize']=(20.0,10.0) #increase the size of graph window
plt.scatter(X,Y, color='Red', label='Data Points')
plt.xlabel('Head Size in Cm')
plt.ylabel('Brain weight in Grams')
plt.legend()
plt.show()
```

```python
#step7: estimating the error or goodness of (line) fit
total_ss = 0
residual_ss = 0
for i in range(n):
    y_prediction = m*X[i]+c
    total_ss += (Y[i]-mean_Y)**2
    residual_ss += (Y[i]-y_prediction)**2
r2 = 1-(residual_ss)/(total_ss)
print(r2)
# grater the r2 max is goodness of line 
```
